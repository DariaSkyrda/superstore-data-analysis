### Outliners

```python
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
for col in num_cols:
    plt.figure(figsize=(8, 3))
    sns.boxplot(x=store_data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

What to do with outliers?

Since the outliers appear to be valid, several approaches are possible: they can be kept as is, capped or floored (winsorization), or analyzed separately from regular orders.

In this analysis, it was decided to retain all outliers without modification to preserve the business relevance of the data.

## Analysis 

### Dataset Info: 
With growing demands and cut-throat competitions in the market, a Superstore Giant is seeking your knowledge in understanding what works best for them. They would like to understand which products, regions, categories and customer segments they should target or avoid.

```python
negative_profit = store_data[store_data['Profit'] < 0]
print(f"Amount of negative orders: {len(negative_profit)}")
print(f"Percentage of negative orders: {(len(negative_profit)/len(store_data))*100:.1f}%")
```

We cannot fully explain the reasons for negative profit because the dataset does not provide information about the profit calculation logic. 
However, negative profit may result from factors such as high discounts, low-margin products, shipping and operational costs, or large quantities sold at a loss.

```python
sns.heatmap(store_data[num_cols].corr(), annot=True, cmap='coolwarm')
plt.show()
```

Profit is positively correlated with sales. Profit is negatively correlated with discount. However, why is discount negatively correlated with sales?

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
sns.barplot(x='Region', y='Sales', data=store_data, ax=axes[0], color="royalblue")
axes[0].set_title("Total Sales by Region")
axes[0].set_xlabel(None)

sns.barplot(x='Region', y='Profit', data=store_data, ax=axes[1], color="lightgreen")
axes[1].set_title("Total Profit by Region")
axes[1].set_xlabel(None)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
sns.barplot(x='Region', y='Sales', data=store_data, ax=axes[0], color="royalblue", hue="Category")
axes[0].set_title("Total Sales by Region and Categories")
axes[0].set_xlabel(None)

sns.barplot(x='Region', y='Profit', data=store_data, ax=axes[1], color="lightgreen", hue="Category")
axes[1].set_title("Total Profit by Region and Categories")
axes[1].set_xlabel(None)

plt.tight_layout()
plt.show()
```

The Central region underperforms in terms of profitability, particularly in the Furniture category and, to a lesser extent, in Office Supplies.
